```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib
```

```python
df=pd.read_csv("zomato.csv",encoding='latin1')
df.head()
```

```python
df.columns
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df_country=pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
country_label=final_df.Country.value_counts().index
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_label[:3],autopct='%.2f')
```

Observation:Maximum orders are from India

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating count'})
ratings
```

```python
matplotlib.rcParams['figure.figsize']=(12,6)
sns.barplot(data=ratings, x='Aggregate rating',y='Rating count',hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(data=ratings,x='Rating color',hue='Rating color',palette=['blue','red','orange','yellow','green','green'])
```
